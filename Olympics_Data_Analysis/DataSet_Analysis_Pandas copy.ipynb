```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<div style="background:#002451;color:#ffffff">First 1000 Records Analysis</div>

```python
# Age Range Analysis of Participants in Olympics over past 120 years.
df = pd.read_csv("athlete_events-1000.csv")

y = list(set(df['Year']))
y.sort()
Max_Age= dict(zip(y,np.zeros(len(y),dtype=int)))
Min_Age= dict(zip(y,np.full(len(y),100,int)))

print(Min_Age)
for i in range(len(df)):
                            # NOTE: Don't use frequent/unwanted access of data values from Data Structures and Dataframes inside Loops, inorder to reduce Execution time. 
    # The below Lines reduces the execution time from 5 Seconds to 0.3 Seconds
    x = df.iloc[i]['Age']               
    st = df.iloc[i]['Year']           
    if(Max_Age[st]<x):
        Max_Age[st]=x
    if(Min_Age[st]>x):
        Min_Age[st]=x
print(Max_Age)
print(Min_Age)

#print((max(Max_Age.keys())-min(Max_Age.keys()))/4)
print(len(list(np.arange(min(Max_Age.keys()),max(Max_Age.keys())+1,2))))
plt.rcParams["figure.figsize"] = [8,6]
plt.rcParams["figure.autolayout"] = True

# Total Medals - Countries Plot
plt.figure(facecolor="#00cccc")          # has to be provided here before the below Statements
ax = plt.axes()
#plt.xticks(rotation=90)
X_axis = np.array(np.arange(len(Max_Age)))
plt.plot(X_axis,list(Max_Age.values()),color="#00ff00",label='Maximum Age')
plt.plot(X_axis,list(Min_Age.values()),color="#ff0000",label='Minimum Age')
ax.set_facecolor("#000000")
plt.title("Plot Analysis of Age Range of Participants in Olympics")
plt.xlabel("Year",size=12,color="#000000")
plt.ylabel("Age",size=12,color="#000000")
plt.xticks(X_axis,Max_Age.keys(),rotation=90)
plt.legend()
plt.show()
```
